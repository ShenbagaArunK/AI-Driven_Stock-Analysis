```python
## LLM and API setup
import os
from langchain_community.utilities import SQLDatabase
from langchain_community.agent_toolkits import create_sql_agent
from langchain_google_genai import ChatGoogleGenerativeAI

key = "sql_12345"
db_url = f'postgresql://postgres:{key}@localhost:5432/stock_analysis_db'
db = SQLDatabase.from_uri(db_url)
print("These are the tables available in the SQL database")
print(db.get_usable_table_names())

## Seting up the LLM with API key -- I am using Gemini LLM
os.environ["GOOGLE_API_KEY"] = "AIzaSyBzX6Yqy0ZDE87GlTgwv_hAvQzOqJFpl7A"

llm = ChatGoogleGenerativeAI(model = "gemini-2.5-flash",temperature=0)
executing_agent = create_sql_agent(llm,db=db,agent_type="tool-calling",verbose=True)

## Testing
response = executing_agent.invoke({"input": "How many stocks are currently in the anchor_50 table? Give me the exact number."})
print("LLM fetched answer")
print(response["output"][0]["text"])
```

```python
## Making the output clean without the signature

raw_data = response["output"]

if isinstance(raw_data,list) and len(raw_data) > 0:
    answer = raw_data[0].get("text", str(raw_data))
elif isinstance(raw_data,str):
    answer = raw_data
else:
    answer = str(raw_data)

print(answer)
```
